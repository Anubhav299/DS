```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('loan_data_set.csv')
```

```python
df
```

```python
df.columns
```

```python
#loan status for gender
pd.crosstab(df['Gender'],df['Loan_Status'])
```

# 1. Loan Status of Gender according to married status.

```python
pd.crosstab([df.Gender,df.Married],df.Loan_Status)
```

# 2.Loan Status on Education Status and Property_Area

```python
pd.crosstab([df.Education,df.Property_Area],df.Loan_Status)
```

# 3.Average Applicant Income according to Education and Sorting according to Gender.

```python
pd.crosstab([df.Education],[df.Gender],values=df.ApplicantIncome,aggfunc=np.average)
```

# 4.Percentage of Loan given according to Credit_History

```python
dt=pd.crosstab(df.Credit_History,df.Loan_Status)
```

```python
dt
```

```python
dt/dt.sum()*100
```

# 5. Percentage of Loan Status according to married status and Dependents.

```python
dt1=pd.crosstab([df.Dependents,df.Married],df.Loan_Status)
```

```python
dt1
```

```python
dt1/dt1.sum()*100
```
