```python
from datetime import datetime

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

data = pd.read_csv('SPY.csv', index_col=0, parse_dates=True)
spx = data['Close']

spx.plot(ax=ax, style='k-')

crisis_data = [(datetime(2007, 10, 8), 'Peak of bull market'),(datetime(2008, 3, 10), 'Bear Stearns Fails'),(datetime(2008, 9, 15), 'Lehman Bankruptcy')]

for date, label in crisis_data:
    ax.annotate(label, xy=(date, spx.asof(date)+5),
                xytext=(date, spx.asof(date)+60),
                arrowprops=dict(facecolor='black', headwidth=10,width=4,headlength=7),
                horizontalalignment='left', verticalalignment='top')

ax.set_title('Important dates in the 2008-2009 financial crisis')
ax.set_xlim(['1/1/2007', '30/9/2009'])
ax.set_ylim([50, 220])
;
```

```python
from datetime import datetime

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

data = pd.read_csv('SPYD.csv', index_col=0, parse_dates=True)
spx = data['Close']

spx.plot(ax=ax, style='k-')

crisis_data = [(datetime(2007, 10, 8), 'Peak of bull market'),(datetime(2008, 3, 10), 'Bear Stearns Fails'),(datetime(2008, 9, 15), 'Lehman Bankruptcy')]

for date, label in crisis_data:
    ax.annotate(label, xy=(date, spx.asof(date)+5),
                xytext=(date, spx.asof(date)+60),
                arrowprops=dict(facecolor='black', headwidth=10,width=4,headlength=7),
                horizontalalignment='left', verticalalignment='top')

ax.set_title('Important dates in the 2008-2009 financial crisis')
ax.set_xlim(['7/7/2007', '7/7/2009'])
ax.set_ylim([60, 220])
;
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

rect = plt.Rectangle((0.2, 0.75), 0.4, 0.15, color='k', alpha=0.3)
circ = plt.Circle((0.7, 0.2), 0.15, color='b', alpha=0.3)
pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]],
 color='g', alpha=0.5)

ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon);
plt.savefig('figpath.svg')
plt.savefig('figpath.png', dpi=400, bbox_inches='tight')
```
