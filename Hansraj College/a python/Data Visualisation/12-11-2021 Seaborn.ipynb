```python
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
```

```python
tips
```

```python
tips.describe()
```

```python
tips.head()
```

```python
#Histogram univariate
sns.distplot(tips['total_bill'],bins=20,kde=False);
```

```python
sns.distplot(tips['tip'],bins=20,kde=False);
```

```python
#Jointplot bivariate scatter
sns.jointplot(x='total_bill',y='tip',data=tips);
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex');
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde');
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg');
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex');
```

```python
sns.pairplot(tips,hue='smoker')
```

```python
#Rugplot
sns.rugplot(tips['total_bill'])
```

```python
#Barplot Categorical
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
import numpy as np


sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
#Boxplot on basis of category
sns.boxplot(x='sex',y='total_bill', data=tips, hue='time');

```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```
